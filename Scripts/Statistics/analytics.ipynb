```python
import pandas as pd
import matplotlib.pyplot as plt
import collections
import numpy as np

# Exploratory Analysis
# Generate beta barrels plot statics

# Aminoacid_Composition:Alanine-A;Aminoacid_Composition:Arginine-R;Aminoacid_Composition:Asparagine-N;Aminoacid_Composition:Aspartiv Acid-D;Aminoacid_Composition:Cysteine-C;Aminoacid_Composition:Glumanine-Q;Aminoacid_Composition:Glutamic acid-E;Aminoacid_Composition:Glycine-G;Aminoacid_Composition:Histidine-H;Aminoacid_Composition:Isoleucine-I;Aminoacid_Composition:Leucine-L;Aminoacid_Composition:Lysine-K;Aminoacid_Composition:Methionine-M;Aminoacid_Composition:Phenylalanine-F;Aminoacid_Composition:Proline-P;Aminoacid_Composition:Serine-S;Aminoacid_Composition:Threonine-T;Aminoacid_Composition:Tryptophan-W;Aminoacid_Composition:Tyrosine-Y;Aminoacid_Composition:Valine-V;Aminoacid_Occurence:Alanine-A;Aminoacid_Occurence:Arginine-R;Aminoacid_Occurence:Asparagine-N;Aminoacid_Occurence:Aspartiv Acid-D;Aminoacid_Occurence:Cysteine-C;Aminoacid_Occurence:Glumanine-Q;Aminoacid_Occurence:Glutamic acid-E;Aminoacid_Occurence:Glycine-G;Aminoacid_Occurence:Histidine-H;Aminoacid_Occurence:Isoleucine-I;Aminoacid_Occurence:Leucine-L;Aminoacid_Occurence:Lysine-K;Aminoacid_Occurence:Methionine-M;Aminoacid_Occurence:Phenylalanine-F;Aminoacid_Occurence:Proline-P;Aminoacid_Occurence:Serine-S;Aminoacid_Occurence:Threonine-T;Aminoacid_Occurence:Tryptophan-W;Aminoacid_Occurence:Tyrosine-Y;Aminoacid_Occurence:Valine-V;Aminoacid_Physico_Composition:Charged(DEKHR);Aminoacid_Physico_Composition:Aliphatic(ILV);Aminoacid_Physico_Composition:Aromatic(FHWY);Aminoacid_Physico_Composition:Polar(DERKQN);Aminoacid_Physico_Composition:Neutral(AGHPSTY);Aminoacid_Physico_Composition:Hydrophobic(CFILMVW);Aminoacid_Physico_Composition:+charged(KRH);Aminoacid_Physico_Composition:-charged(DE);Aminoacid_Physico_Composition:Tiny(ACDGST);Aminoacid_Physico_Composition:Small(EHILKMNPQV);Aminoacid_Physico_Composition:Large(FRWY);Aminoacid_Physico_Occurence:Charged(DEKHR);Aminoacid_Physico_Occurence:Aliphatic(ILV);Aminoacid_Physico_Occurence:Aromatic(FHWY);Aminoacid_Physico_Occurence:Polar(DERKQN);Aminoacid_Physico_Occurence:Neutral(AGHPSTY);Aminoacid_Physico_Occurence:Hydrophobic(CFILMVW);Aminoacid_Physico_Occurence:+charged(KRH);Aminoacid_Physico_Occurence:-charged(DE);Aminoacid_Physico_Occurence:Tiny(ACDGST);Aminoacid_Physico_Occurence:Small(EHILKMNPQV);Aminoacid_Physico_Occurence:Large(FRWY);AA;AR;AN;AD;AC;AQ;AE;AG;AH;AI;AL;AK;AM;AF;AP;AS;AT;AW;AY;AV;RA;RR;RN;RD;RC;RQ;RE;RG;RH;RI;RL;RK;RM;RF;RP;RS;RT;RW;RY;RV;NA;NR;NN;ND;NC;NQ;NE;NG;NH;NI;NL;NK;NM;NF;NP;NS;NT;NW;NY;NV;DA;DR;DN;DD;DC;DQ;DE;DG;DH;DI;DL;DK;DM;DF;DP;DS;DT;DW;DY;DV;CA;CR;CN;CD;CC;CQ;CE;CG;CH;CI;CL;CK;CM;CF;CP;CS;CT;CW;CY;CV;QA;QR;QN;QD;QC;QQ;QE;QG;QH;QI;QL;QK;QM;QF;QP;QS;QT;QW;QY;QV;EA;ER;EN;ED;EC;EQ;EE;EG;EH;EI;EL;EK;EM;EF;EP;ES;ET;EW;EY;EV;GA;GR;GN;GD;GC;GQ;GE;GG;GH;GI;GL;GK;GM;GF;GP;GS;GT;GW;GY;GV;HA;HR;HN;HD;HC;HQ;HE;HG;HH;HI;HL;HK;HM;HF;HP;HS;HT;HW;HY;HV;IA;IR;IN;ID;IC;IQ;IE;IG;IH;II;IL;IK;IM;IF;IP;IS;IT;IW;IY;IV;LA;LR;LN;LD;LC;LQ;LE;LG;LH;LI;LL;LK;LM;LF;LP;LS;LT;LW;LY;LV;KA;KR;KN;KD;KC;KQ;KE;KG;KH;KI;KL;KK;KM;KF;KP;KS;KT;KW;KY;KV;MA;MR;MN;MD;MC;MQ;ME;MG;MH;MI;ML;MK;MM;MF;MP;MS;MT;MW;MY;MV;FA;FR;FN;FD;FC;FQ;FE;FG;FH;FI;FL;FK;FM;FF;FP;FS;FT;FW;FY;FV;PA;PR;PN;PD;PC;PQ;PE;PG;PH;PI;PL;PK;PM;PF;PP;PS;PT;PW;PY;PV;SA;SR;SN;SD;SC;SQ;SE;SG;SH;SI;SL;SK;SM;SF;SP;SS;ST;SW;SY;SV;TA;TR;TN;TD;TC;TQ;TE;TG;TH;TI;TL;TK;TM;TF;TP;TS;TT;TW;TY;TV;WA;WR;WN;WD;WC;WQ;WE;WG;WH;WI;WL;WK;WM;WF;WP;WS;WT;WW;WY;WV;YA;YR;YN;YD;YC;YQ;YE;YG;YH;YI;YL;YK;YM;YF;YP;YS;YT;YW;YY;YV;VA;VR;VN;VD;VC;VQ;VE;VG;VH;VI;VL;VK;VM;VF;VP;VS;VT;VW;VY;VV;
# #RelatedDomains;#PhobiusAlfa;PhobiusSignalPeptide;#BompBetaBarrels;SubcellularLoc;Transporter


def format_labels(val):
    return '{:.2f}%\n({:.0f})'.format(val, total_sequences*val/100)

def format_labels_transporters(val):
    return '{:.2f}%\n({:.0f})'.format(val, number_transporters*val/100)

# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
        
data = pd.read_csv('/Users/marcelo/Desktop/Rita/Tese/tese/Data/Infeatures/InFeatures_data-filtered-random-final.csv', sep=';')
total_sequences = len(data.index)

transporters = data.get(data['Transporter'] == 1)
number_transporters = len(transporters.index)
number_non_transporters = total_sequences - number_transporters

PIE_DPI = 150
HIST_DPI = 100

# Draw transporters chart
transporters_labels = ["No transportadores", "Transportadores"]
transporters_values = [number_non_transporters, number_transporters]

plt.title('Taxa Transportadores')
plt.pie(transporters_values, labels=transporters_labels, autopct=format_labels)
plt.savefig("transporters.png", dpi=PIE_DPI, transparent=True)
plt.show()

# Draw beta barrels chart
beta_barrels_counter = collections.Counter(data["#BompBetaBarrels"].tolist())
beta_barrels_values = list(beta_barrels_counter.values())

beta_barrels_labels = ["No Beta", "Beta"]

plt.title('Taxa Beta Barrels')
plt.pie(beta_barrels_values, labels=beta_barrels_labels, autopct=format_labels)
plt.savefig("beta_barrels.png", dpi=PIE_DPI, transparent=True)
plt.show()

# Draw beta barrels chart - transporters
beta_barrels_transporter_counter = collections.Counter(transporters["#BompBetaBarrels"].tolist())
beta_barrels_transporter_values = list(beta_barrels_transporter_counter.values())

plt.title('Taxa Beta Barrels - Transportadores')
plt.pie(beta_barrels_transporter_values, labels=beta_barrels_labels, autopct=format_labels_transporters)
plt.savefig("beta_barrels_transporters.png", dpi=PIE_DPI, transparent=True)
plt.show()

# Draw Signal peptide
signal_counter = collections.Counter(data["PhobiusSignalPeptide"].tolist())
signal_values = list(signal_counter.values())

signal_labels = ["Sem peptido sinal", "Peptido sinal"]

plt.title('Taxa Peptido sinal')
plt.pie(signal_values, labels=signal_labels, autopct=format_labels)
plt.savefig("peptido_sinal.png", dpi=PIE_DPI, transparent=True)
plt.show()

# Draw Signal peptide - transporters
signal_transporter_counter = collections.Counter(transporters["PhobiusSignalPeptide"].tolist())
signal_transporter_values = list(signal_transporter_counter.values())

plt.title('Taxa Peptido sinal - Transportadores')
plt.pie(signal_transporter_values, labels=signal_labels, autopct=format_labels_transporters)
plt.savefig("peptido_sinal_transportadores.png", dpi=PIE_DPI, transparent=True)
plt.show()

# Draw alfa helixes
alpha_counter = collections.Counter(data["#PhobiusAlfa"].tolist())

not_alpha_counter = alpha_counter[0]

alpha_labels = ["Sem Alfa-Hlices", "Com Alfa-Hlices"]
alpha_values = [not_alpha_counter, total_sequences - not_alpha_counter]

plt.title('Taxa Alfa-Hlices')
plt.pie(alpha_values, labels=alpha_labels, autopct=format_labels)
plt.savefig("alfa.png", dpi=PIE_DPI, transparent=True)
plt.show()

# Draw alfa helixes - transporters
alpha_transporter_counter = collections.Counter(transporters["#PhobiusAlfa"].tolist())
not_alpha_transporter_counter = alpha_transporter_counter[0]

alpha_transporter_values = [not_alpha_transporter_counter, number_transporters - not_alpha_transporter_counter]

plt.title('Taxa Alfa-Hlices - Transportadores')
plt.pie(alpha_transporter_values, labels=alpha_labels, autopct=format_labels_transporters)
plt.savefig("alfa_transportadores.png", dpi=PIE_DPI, transparent=True)
plt.show()

# histogram of alfa helixes
bins = np.array(range(1, 42, 1))
a = []

for value in alpha_counter.most_common():
    if value[0] != 0.0:
        a.extend([value[0]] * value[1])

fig, ax = plt.subplots(figsize = (15, 7))
ax.hist(a, bins = bins)

plt.title("Distribuio de quantidade de Alfa-Hlices")
plt.savefig("alfa_dist.png", dpi=HIST_DPI, transparent=True)
plt.show()

# histogram of alfa helixes - Transportadores
bins = np.array(range(1, 42, 1))
a = []

for value in alpha_transporter_counter.most_common():
    if value[0] != 0.0:
        a.extend([value[0]] * value[1])

fig, ax = plt.subplots(figsize = (15, 7))
ax.hist(a, bins = bins)

plt.title("Distribuio de quantidade de Alfa-Hlices - Transportadores")
plt.savefig("alfa_dist_transportadores.png", dpi=HIST_DPI, transparent=True)
plt.show()

# Draw subcellulalr location
subcellular_loc_counter = collections.Counter(data["SubcellularLoc"].tolist())
subcellular_dictionary = ["NA", "chloroplast", "chloroplast memb.", "cytosol", "endoplasmic \nreticulum", "endoplasmic \nreticulum memb.", "extra-cellular", "mbrium", "golgi \napparatus", "golgi \napparatus memb.", "mitochondrion", "mitochondrion \nmemb.", "nucleus", "nucleus memb.", "outer memb.", "periplasmic space", "peroxisome", "peroxisome memb.", "plasma memb.", "plastid", "vacuole", "vacuole memb.", "secreted", "cytoplasm", "inner memb."]

values = []
values_description = []

for value in subcellular_loc_counter.most_common():
    values.append(value[1])
    values_description.append(subcellular_dictionary[int(value[0])])

fig = plt.figure(figsize = (15, 5))
plt.bar(values_description[:10], values[:10], color ='maroon', width = 0.4)

addlabels(values_description[:10], values[:10])

plt.title("Top 10 - Distribuio de quantidade de localizao celular")
plt.savefig("cel_loc_dist.png", dpi=HIST_DPI, transparent=True)
plt.show()


# Draw subcellulalr location - Transporters
subcellular_loc_counter = collections.Counter(transporters["SubcellularLoc"].tolist())
subcellular_dictionary = ["NA", "chloroplast", "chloroplast memb.", "cytosol", "endoplasmic \nreticulum", "endoplasmic \nreticulum memb.", "extra-cellular", "mbrium", "golgi \napparatus", "golgi apparatus \nmemb.", "mitochondrion", "mitochondrion \nmemb.", "nucleus", "nucleus memb.", "outer memb.", "periplasmic space", "peroxisome", "peroxisome memb.", "plasma memb.", "plastid", "vacuole", "vacuole memb.", "secreted", "cytoplasm", "inner memb."]

values = []
values_description = []

for value in subcellular_loc_counter.most_common():
    values.append(value[1])
    values_description.append(subcellular_dictionary[int(value[0])])

fig = plt.figure(figsize = (15, 5))
plt.bar(values_description[:10], values[:10], color ='maroon', width = 0.4)

addlabels(values_description[:10], values[:10])

plt.title("Top 10 - Distribuio de quantidade de localizao celular - Transportadores")
plt.savefig("cel_loc_dist_transportadores.png", dpi=HIST_DPI, transparent=True)
plt.show()


# Draw most representative dipeptides

def get_max_dipeptides_counter(base_data):

    first_column_index = base_data.columns.get_loc("AA")
    last_column_index = base_data.columns.get_loc("VV")

    dipeptides = base_data.iloc[:, first_column_index:last_column_index+1]
    dipeptides_max_column_by_rows = dipeptides.idxmax(axis=1)

    return collections.Counter(dipeptides_max_column_by_rows.tolist())

def generate_dipeptide_graph(counter, title, filename):
    values = []
    values_description = []

    for value in counter.most_common():
        values.append(value[1])
        values_description.append(value[0])

    fig = plt.figure(figsize = (15, 5))
    plt.bar(values_description[:10], values[:10], color ='maroon', width = 0.4)

    addlabels(values_description[:10], values[:10])

    plt.title(title)
    plt.savefig(filename, dpi=HIST_DPI, transparent=True)
    plt.show()

# generate global dipeptides
max_by_dipeptides_counter = get_max_dipeptides_counter(data)
generate_dipeptide_graph(max_by_dipeptides_counter, "Top 10 - Dipeptidos com maior representatividade", "dipeptidos_repr.png")

# generate transporters dipeptides
max_by_dipeptides_counter_transporters = get_max_dipeptides_counter(transporters)
generate_dipeptide_graph(max_by_dipeptides_counter_transporters, "Top 10 - Dipeptidos com maior representatividade - Transportadores", "dipeptidos_repr_transportadores.png")
```
